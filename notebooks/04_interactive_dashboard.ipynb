```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("../data/master_econ_data.csv")
df["date"] = pd.to_datetime(df["date"])
```

```python
fig = px.line(df, x="date", y="cpi", title="CPI (2024–Today)")
fig.show()
```

```python
fig = go.Figure()

# Add all indicators as separate traces
fig.add_trace(go.Scatter(x=df["date"], y=df["cpi"], name="CPI", visible=True))
fig.add_trace(go.Scatter(x=df["date"], y=df["unemployment_rate"], name="Unemployment Rate", visible=False))
fig.add_trace(go.Scatter(x=df["date"], y=df["gdp"], name="GDP", visible=False))
fig.add_trace(go.Scatter(x=df["date"], y=df["fed_funds_rate"], name="Fed Funds Rate", visible=False))

# Dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(label="CPI",
                     method="update",
                     args=[{"visible": [True, False, False, False]}]),
                dict(label="Unemployment Rate",
                     method="update",
                     args=[{"visible": [False, True, False, False]}]),
                dict(label="GDP",
                     method="update",
                     args=[{"visible": [False, False, True, False]}]),
                dict(label="Fed Funds Rate",
                     method="update",
                     args=[{"visible": [False, False, False, True]}]),
            ]),
            direction="down"
        )
    ],
    title="Economic Indicators (Interactive Dashboard)",
    xaxis_title="Date",
    yaxis_title="Value"
)

fig.show()
```

```python
df["cpi_norm"] = df["cpi"] / df["cpi"].iloc[0]
df["unemp_norm"] = df["unemployment_rate"] / df["unemployment_rate"].iloc[0]
df["gdp_norm"] = df["gdp"] / df["gdp"].iloc[0]
df["fed_norm"] = df["fed_funds_rate"] / df["fed_funds_rate"].iloc[0]

df["economic_health_index"] = (
    df["gdp_norm"] * 0.4 +
    (1 / df["unemp_norm"]) * 0.3 +
    (1 / df["cpi_norm"]) * 0.2 +
    (1 / df["fed_norm"]) * 0.1
)
```

```python
fig = px.line(df, x="date", y="economic_health_index",
              title="Economic Health Index (2024–Today)")
fig.show()
```
